```python
import numpy as np
import matplotlib.pyplot as plt

# Define the dimensions of the image
width = 160
height = 160

# Open the .bin file and read the data
with open('pre_st_img.bin', 'rb') as f:
    data = f.read()

# Convert the data to a numpy array
pixels = np.frombuffer(data, dtype=np.uint8)

# Reshape the array into an image
image = pixels.reshape((height, width, 3))

# Display the image
plt.imshow(image)
plt.show()
```
